```python
url='/content/mtcars.csv'
data= pd.read_csv(url)
data.hist()
plt.show()
correlation= data.corr()
print(correlation)
X=data.drop(data.columns[[0,2]],axis=1)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='grey')
```

```python
from sklearn import  linear_model# import thu vien sklearn nham de machien tu hoc duoc, voi phan linear_model nham de the hien duong phan linear_model
model=linear_model.LinearRegression()#tao ham hi quy tuyen tinh
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg'])#nhap cac du lieu cua data bao gom "wt" va "mpg"
print(model.intercept_)#gia tri trung binh khi ta cho cac bien du doan cua ham hoi quy tuyen tinh bang khong.
print(model.coef_)

print(model.score(pd.DataFrame(data['wt']),data['mpg']))
y_pred= model.predict(pd.DataFrame(data['wt']))
residual= data['mpg'] - y_pred#sai so giua du lieu goc va du lieu duoc du doan
residual.describe()
print(data['mpg'],end=' ')
print(y_pred,end=' ')
print(residual,end=' ')

data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='red')
plt.plot(data['wt'],y_pred,color='blue')
```

```python

model=linear_model.LinearRegression()
X= pd.DataFrame([data['wt'],data['wt']**2]).T
y=data['mpg']
model.fit(X,y=data['mpg'])
print("Intercpet of model two degrees:")
print(model.intercept_)
print('Cofienct of model two degrees:')
print(model.coef_)
print('The accuracy of the model')
print(model.score(X,y))
#--------------------------
range=np.arange(1.5,5.5,0.1)#do phan wt tropng du lieu chi trong khoang tu [0,5.5] nen ta se su dung mot ham range voi khoang nhu vat
predictor=pd.DataFrame([range,range**2]).T
y_pred= model.predict(X=predictor)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```

```python
import seaborn as sns
data=pd.read_csv("/content/50_Startups.csv")
data.head()

sns.pairplot(data)
X=data.drop('Profit',axis=True)#delete of column Profit values
X=data.drop('State',axis=True)#delete one more column State cause it is string
y=data['Profit']
X.head()

from sklearn.preprocessing import StandardScaler#Scale cac so nho len thanh cac gia tri phu hop cho may hoc
sc=StandardScaler()
X=sc.fit_transform(X)

from sklearn.model_selection import train_test_split# chia du lieu thanh cac phan train va test


X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
corr=data.corr()
sns.heatmap(corr,annot=True)
```

```python

from sklearn.linear_model import LinearRegression
#Starting machine learning

regressor=LinearRegression()
regressor.fit(X_train,y_train)
y_pred= regressor.predict(X_test).round(1)#prediction with rouding one num after comma
calculation=pd.DataFrame(np.c_[y_test,y_pred],columns=['Original Profits','Predicted Profit'])

print(regressor.score(X_test, y_test))

print(regressor.score(X_train,y_train))
```
